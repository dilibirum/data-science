# Анализ госзакупок по запросу покупок лицензий Битрикс24 

```python
import pandas as pd
import unicodedata
import sys
```

```python
tenders = pd.read_excel('data/bitrix24.xlsx')
```

```python
tenders['customer'] = tenders['customer'].apply(lambda s: s.lower())
```

```python
tenders = tenders.dropna()
```

```python
stopwords = {
    '"': None,
    'общество': None,
    'с': None,
    'ограниченной': None,
    'ответственностью': None,
    'ооо': None,
    'индивидуальный': None,
    'предприниматель': None,
    'ип': None,
}
```

```python
tenders['vendor'] = tenders['vendor'].apply(lambda s: s.lower().translate(stopwords))
```
