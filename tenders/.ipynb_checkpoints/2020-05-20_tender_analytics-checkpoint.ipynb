```python
def to_float(string):
    res = np.nan
    try:
        numbers = string.split(',')
        integer = int(numbers[0])
        fractional = int(numbers[1])
        res = integer + fractional / 10 ** len(numbers[1])
    except:
        pass
    return res
```

```python
def get_short_name(string):
    try:
        res = string.split('"')
        res = ' '.join(map(lambda s: s.lower().capitalize(), res[1].split()))
    except IndexError:
        res = ' '.join(map(lambda s: s.lower().capitalize(), string.split()))
    return res
```

```python
avrs = dict()
for year in range(2012, 2021):
    d = aa.loc[(year, )]['amount']
    average = d.sum() / d.count()
    avrs[year] = average
#     means.append(avrs)
```

```python
def barplot(sample: pd.Series, 
            title='', 
            xlabel='', 
            ylabel='', 
            has_labels=False, 
            has_stat=False, 
            is_rotate=False,
            has_mean=False,
            orientation='v',
            
           ):
    
    fig, ax = plt.subplots()
    plt.title(title, fontsize=14)
    
    if orientation == 'v':
        if has_mean:
            plt.axhline(y=sample.values.mean(), color=(0.25, 0.5, 0.75), linestyle='-')
            plt.text(x=0.05, 
                     y=sample.values.mean() * 1.15, 
                     s=f"В среднем: {sample.values.mean():.0f}", 
                     ha='left', 
                     va='center', 
                     color='black', 
                     fontdict={'fontsize': 14})

        if has_labels:
            for x, y in enumerate(sample.values):
                plt.text(x=x,
                         y=y + sample.values.max() * 0.05,
                         s=f"{y:,.0f} ({y / sample.values.sum():.1%})".replace(',', ' '),
                         ha='center',
                         fontdict={'fontsize': 10}
                )
                
        if has_stat:
            for i, year in zip(range(3, 90, 12), range(2013, 2021)):
                plt.axvline(x=i, color=(0.25, 0.5, 0.75), linestyle=':')
                plt.text(x=i + 0.5, y=262.5, s=str(year), ha='left', va='center', 
                         color=(0.25, 0.5, 0.75), fontdict={'fontsize': 12})

                plt.hlines(y=avrs[year],
                           xmin=i,
                           xmax=i + 12,
                           color=(0.25, 0.5, 0.75),
                           linestyles='dashed',
                          )
                if year != 2020:
                    plt.text(x=i + 6, y=avrs[year] * 1.1, s=f'{avrs[year]:.0f}', ha='center', va='bottom', 
                             color='black', fontdict={'fontsize': 12})
                else:
                    plt.text(x=i + 1.5, y=avrs[year] * 1.1, s=f'{avrs[year]:.0f}', ha='left', va='bottom', 
                             color='black', fontdict={'fontsize': 12})

            plt.hlines(y=avrs[2012],
                       xmin=0,
                       xmax=3,
                       color=(0.25, 0.5, 0.75),
                       linestyles='dashed',
                       label='',
                      )

            plt.text(x=1.25, y=avrs[2012] * 1.1, s=f'{avrs[2012]:.0f}', ha='center', va='bottom', 
                             color='black', fontdict={'fontsize': 12})

        fig = sns.barplot(x=sample.index.astype(str),
                    y=sample.values,
                    )
        if is_rotate:
            plt.xticks(rotation=45, ha='right', size=8)
        else:
            plt.xticks(rotation=0, ha='center', size=8)

        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel);
    
    elif orientation == 'h':
        fig = sns.barplot(y=sample.index.astype(str),
                          x=sample.values,
                          orient='h',
                          )
        if has_labels:
            for y, x in enumerate(sample.values):
                plt.text(x=x + sample.values.max() * 0.01,
                         y=y,
                         s=f"{x:,.0f} ({x / sample.values.sum():.1%})".replace(',', ' '),
                         va='center',
                         fontdict={'fontsize': 10}
                )
        
```

```python
def pieplot(sample, title=''):
    
    def func(pct, allvals):
        absolute = int(pct / 100 * np.sum(allvals))
        return "{:.1f}%\n({:,.2f})".format(pct, absolute).replace(',', ' ')
    
    plt.title(title, fontsize=14)


    plt.pie(sample,
            labels=sample.index,
            autopct=lambda pct: func(pct, sample.values),
           );

    plt.legend(bbox_to_anchor=(1, 0, 0.5, 1));
```

###### Анализ ключевых слов для запроса "Разработка ПО"

```python
with open('data/stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.readlines()
stop_words_upd = [st[:-1] for st in stop_words]
```

```python
stop_words_upd += ['"\n', ')\n', ' \n', '\n', '»\n', 'предложение', 
                   'нужда', 'оказание', 'работа', 'услуга', 'запрос',
                   'группа', 'закупка', 'оао', 'выбор', 'часть',
                  ]
```

```python
def key_words(data):
    w = data['description'].apply(m.lemmatize)
    words = []
    for word in w:
        words += word
    words_upd = [word for word in words if word not in stop_words_upd]
    return Counter(words_upd)
```

```python
%%time

sw_key_words = key_words(df.query('query == "разработка ПО"'))
```

```python
%%time

is_key_words = key_words(df.query('query == "разработка ИС"'))
```

```python
sw_top_100_words = sw_key_words.most_common()[10:110]
sw_top_10_words = sw_key_words.most_common()[:10]
```

```python
def tag_cloud(text, title='', fname=''):
        
    wordcloud = WordCloud(width=1600, 
                          height=1000, 
                          background_color="white", 
                          relative_scaling=1.0,
                          repeat=True, 
                          collocations=False, 
                          min_font_size=8,
                         ).generate_from_frequencies(dict(text))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=20)
    plt.axis("off")
    plt.figure(figsize=(16, 10))
    plt.tight_layout()
    wordcloud.to_file(fname)
    plt.show();
```

### Топ заказчиков

```python
authors = df.query('price > 0') \
            .pivot_table(index='authors_short_name',
                           values=['id', 'price'],
                           aggfunc={'id': 'count', 'price': 'sum'}
                           ) \
            .rename(columns={'id': 'tenders'})
authors['mean_price'] = authors['price'] / authors['tenders']
```

```python
authors_top_by_tender_amount = authors.sort_values(by='tenders', ascending=False)
authors_top_by_tender_amount[['tenders']].head(10)
```

```python
authors_top_by_price = authors.sort_values(by='price', ascending=False)
authors_top_by_price[['price']].head(10)
```

```python
authors_top_by_mean_price = authors.sort_values(by='mean_price', ascending=False)
authors_top_by_mean_price[['mean_price']].head(10)
```

```python
def get_top_authors(limit):
    s1 = set(authors_top_by_tender_amount[:limit].index) & set(authors_top_by_price[:limit].index)
    s2 = set(authors_top_by_price[:limit].index) & set(authors_top_by_mean_price[:limit].index)
    s3 = set(authors_top_by_tender_amount[:limit].index) & set(authors_top_by_mean_price[:limit].index)

    return s1.union(s2).union(s3)
```

```python
top_authors = get_top_authors(15)
top_authors
```

```python
tenders_pct = authors_top_by_tender_amount.query('authors_short_name in @top_authors')['tenders'].sum() \
              / df.shape[0]
authors_pct = len(top_authors) / df['author'].nunique()
price_pct = authors_top_by_price.query('authors_short_name in @top_authors')['price'].sum() / df['price'].sum()
mean_price = authors_top_by_mean_price.query('authors_short_name in @top_authors')['price'].sum() \
            / authors_top_by_mean_price.query('authors_short_name in @top_authors')['tenders'].sum()
```

```python
print(f"Топ {len(top_authors)} заказчиков составляют {authors_pct:.1%} от общего числа,"
      + f" \nразмещают {tenders_pct:.1%} тендеров " 
      + f"и занимают {price_pct:.1%} рынка гос. закупок по объему заказов")
print(f"Средний размер тендера: {mean_price:,.2f} ₽")
```
