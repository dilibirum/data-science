Согласно Википедии, выброс - это такая точка наблюдения, которая удалена от других наблюдений. Это определение расплывчато, потому что оно не дает количественной оценки слову "удалена". В этом статье мы попытаемся разобраться в различных интерпретациях понятия "удалена". А также мы рассмотрим методы обнаружения и обработки выбросов, наблюдая их влияние на различные типы моделей машинного обучения.

Выбросы возникают из-за изменений в поведении системы, мошенничества, человеческого фактора, ошибки приборов или просто из-за естественных отклонений в данных. Образец данных мог быть загрязнен значениями, поступившими из-за пределов исследуемой области.

Многие модели машинного обучения, такие как линейная и логистическая регрессия, легко подвержены влиянию выбросов в обучающих данных. Такие модели, как AdaBoost, увеличивают вес неправильно классифицированных точек на каждой итерации и поэтому могут устанавливать высокие веса для этих выбросов, поскольку они часто неправильно классифицируются. Это может стать проблемой, если этот выброс является ошибкой какого-то типа, или если мы хотим, чтобы наша модель хорошо обобщала и не заботилась об экстремумах.

Чтобы преодолеть эту проблему, мы можем либо изменить модель или метрику, либо внести некоторые изменения в данные и использовать одни и те же модели. Для анализа мы рассмотрим данные о ценах на жилье Kaggle: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data 


### Что мы подразумеваем под выбросами?
Экстремальные значения могут присутствовать как в зависимых, так и в независимых переменных, в случае контролируемых методов обучения. 

Эти экстремальные значения не обязательно влияют на производительность или точность модели, но когда это происходит, они называются "влияющими" точками.

#### Экстремальные значения независимых переменных
Такие значения называются точками ***"высокого рычага"*** (high leverage). С единственным предиктором экстремальное значение — это просто то, которое очень высоко или низко. При наличии нескольких предикторов экстремальные значения могут быть очень высокими или низкими для одного или нескольких предикторов (***одномерный анализ (univariate analysis) - анализ одной переменной единовременно***) или могут быть комбинациями значений предикторов ***(многомерный анализ – multivariate analysis)***.

На следующем рисунке все точки с правой стороны оранжевой линии являются точками высокого рычага.

#### Экстремальные значения целевых переменных
Для задач регрессии - эти экстремальные значения называются "выбросами". Они могут как быть так и не быть влияющими точками, что мы увидим позже. На следующем рисунке все точки над оранжевой линией можно классифицировать как выбросы.


```python
Image('scr/img/how-pic 2.png')
```

Для задач классификации можно выделить два типа экстремальных значений:

1) **Выбросы**: например, в задаче классификации изображений, в которой мы пытаемся идентифицировать собак/кошек, одно из изображений в обучающем наборе является изображением гориллы (или любым другым изображением, не являющимся частью задачи) по ошибке. Здесь изображение гориллы – это явно шум. Обнаружение выбросов здесь не имеет смысла, потому что мы уже знаем, на каких категориях мы хотим сосредоточиться, а какие отбросить.

2) **Новшества**: мы много раз имеем дело с новшествами, и эта проблема часто называется **контролируемым обнаружением аномалий**. В данном случае цель состоит не в том, чтобы устранить выбросы или уменьшить их влияние, а в том, чтобы выявить аномалии в новых наблюдениях. Поэтому мы не будем обсуждать это в этой статье. Новшества особенно используются для обнаружения мошенничества при операциях с кредитными картами, фальшивых звонках и т. д.

Все пункты, которые мы обсуждали выше, включая влияющие точки, хорошо иллюстрирует следующий рисунок:

```python
Image('scr/img/how-pic 3.png')
```

где:
* точки в области Q1 – это выбросы (outliers);
* точки в области Q3 – это точки высокого рычага (leverage points);
* точки в области Q2 – это и выбросы и точки высокго рычага, но не влияющие точки (influential points);
* обведенные точки – пример влияющих точек. Их может быть и больше, но эти самые яркие их представители.

Наше основное внимание будет сосредоточено на выбросах (экстремальных значениях целевой переменной для дальнейшего исследования). Мы увидим влияние этих экстремальных значений на производительность модели.
